## Customer who exceed the limit in any particluar month

## Which age group is spending more money?

```python
age_range = [18,25,30,35,40,45,50,55,60,65,70,75,80]
merge_credit_spend['Age Group'] = pd.cut(merge_credit_spend['Age'], bins=age_range, labels=['18-25','25-30','30-35','35-40','40-45','45-50','50-55','55-60','60-65','65-70','70-75','75-80'])
age_group_spending =  merge_credit_spend.groupby(['Age Group','S Month', 'S Year'])['Amount'].sum()
```

```python
age_group_spending = pd.DataFrame(age_group_spending)
display_full_dataframe(age_group_spending)
```
